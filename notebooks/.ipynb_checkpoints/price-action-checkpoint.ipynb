# Initialize

```python
# Import pandas before using it
import pandas as pd
import numpy as np  # Add this import for using np.where
```

```python
select c.*, mc.symbol,mc.interval from Candle c left join MetaCandle mc on mc.id=c.metaCandleId;
```

```python
# Fixing 'openTime' and 'closeTime' conversion to dates
# Convert openTime to datetime only if the values are numerical, otherwise directly convert them to date
candle_df = df_5

candle_df['open_date'] = pd.to_datetime(df_5['openTime'], unit='ms').dt.date  # Convert openTime to Date
candle_df['close_date'] = pd.to_datetime(df_5['closeTime'], unit='ms').dt.date  # Convert closeTime to Date

# Add trend column using np.where for conditional logic
candle_df['trend'] = np.where(candle_df['open'] > candle_df['close'], "bullish", "bearish")

# Add additional calculated columns
candle_df['high_low_percent'] = round((candle_df['high'] - candle_df['low']) / candle_df['low'] * 100)
candle_df['high_open_percent'] = round((candle_df['high'] - candle_df['open']) / candle_df['open'] * 100)
candle_df['high_close_percent'] = round((candle_df['high'] - candle_df['close']) / candle_df['close'] * 100)
candle_df['open_low_percent'] = round((candle_df['open'] - candle_df['low']) / candle_df['low'] * 100)
candle_df['close_low_percent'] = round((candle_df['close'] - candle_df['low']) / candle_df['low'] * 100)

candle_df
```

### Join weekly data with daily

```python
# Add a join column for mapping '1d' interval rows to '1w' interval rows based on symbol and time conditions

# Separate the DataFrame into '1d' and '1w' intervals for processing
df_1d = candle_df[candle_df['interval'] == '1d']
df_1w = candle_df[candle_df['interval'] == '1w']

# Perform a merge (self-join) based on symbol and time conditions
merged_df = pd.merge(
    df_1d, 
    df_1w, 
    how='left', 
    left_on='symbol', 
    right_on='symbol', 
    suffixes=('_1d', '_1w')
)

# Apply the time conditions for the join
merged_df = merged_df[
    (merged_df['openTime_1d'] >= merged_df['openTime_1w']) &
    (merged_df['closeTime_1d'] <= merged_df['closeTime_1w'])
]

# Add 'join_column' to the original '1d' DataFrame with the corresponding 1w 'id'
# Use the first matching 'id' in case of multiple matches
df_1d = df_1d.merge(
    merged_df[['id_1d', 'id_1w']].drop_duplicates(subset=['id_1d']), 
    how='left', 
    left_on='id', 
    right_on='id_1d'
)

# Rename and clean up the join column
df_1d = df_1d.rename(columns={'id_1w': 'join_column'}).drop(columns=['id_1d'])

# Combine the modified '1d' rows back into the original DataFrame
candle_df = pd.concat([df_1d, candle_df[candle_df['interval'] != '1d']], ignore_index=True)

candle_df
```

# Visualize

### this is a function to draw charts

```python
import matplotlib.pyplot as plt

def plot_symbol_chart(candle_df, symbol: str, interval: str, tail_count: int = 40):
    # Filter the DataFrame
    df = candle_df[(candle_df['symbol'] == symbol) & (candle_df['interval'] == interval)].copy()

    # Sort by open_date and keep the latest `tail_count` rows
    df = df.sort_values(by='open_date').tail(tail_count)

    # محاسبه نسبت volume به numberOfTrades
    df['volume_per_trade'] = df['volume'] / df['numberOfTrades']

    # Create the figure and first axis
    fig, ax1 = plt.subplots(figsize=(16, 8))

    # Map trend to colors
    colors = df['trend'].map({'bullish': 'green', 'bearish': 'red'})
    ax1.bar(df['open_date'], df['high_low_percent'], color=colors, label='High-Low Percent')

    # Configure left axis
    ax1.set_xlabel('Open Date')
    ax1.set_ylabel('High-Low Percent')
    ax1.set_title(f'{symbol} High-Low Percent and Volume ({interval} Interval)')
    ax1.tick_params(axis='x', rotation=45)

    # Create the second axis for volume
    ax2 = ax1.twinx()
    ax2.plot(df['open_date'], df['volume_per_trade'], color='blue', marker='o', label='Volume per Trade')
    ax2.set_ylabel('Volume', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')

    # Add legends
    ax1.legend(loc='upper left', title='Bar Chart')
    ax2.legend(loc='upper right', title='Line Plot')

    # Add bullish/bearish count box
    bullish_count = (df['trend'] == 'bullish').sum()
    bearish_count = (df['trend'] == 'bearish').sum()
    textstr = f'Bullish: {bullish_count}\nBearish: {bearish_count}'
    props = dict(boxstyle='round', facecolor='lightgrey', alpha=0.7)
    ax1.text(0.45, 0.95, textstr, transform=ax1.transAxes, fontsize=12,
             verticalalignment='top', bbox=props)

    # Clean layout and show plot
    fig.tight_layout()
    plt.show()
```

```python
plot_symbol_chart(candle_df,'BTCUSDT', '1w')
```
